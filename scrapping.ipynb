```python
prices=[]
N=[]
classes = []
depart=[]
arraival=[]
Depature_Place = []
Arraival_Place = []
time =[]
stop = []
lag = []

for url in urls: 
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    container = soup.find_all("div", class_="yR1fYc")
    
    for i in container:
        Price = i.find("div", class_="U3gSDe ETvUZc")
        if Price:
            prices.append(Price.text)
        else:
            prices.append('price unavailable')
        
        name = i.find("div",class_="sSHqwe tPgKwe ogfYpf")
        if name:
            N.append(name.text)
        
        departures = i.find("div", class_="wtdjmc YMlIz ogfYpf tPgKwe")
        if departures:
            depart.append(departures.text)
        
        arrivals = i.find("div", class_="XWcVob YMlIz ogfYpf tPgKwe")
        if arrivals:
            arraival.append(arrivals.text.split("(")[0])
            
        D_P = i.find("div", class_="G2WY5c sSHqwe ogfYpf tPgKwe")
        if D_P:
            Depature_Place.append(D_P.text)
        
        A_P = i.find("div",class_="c8rWCd sSHqwe ogfYpf tPgKwe")
        if A_P:
            Arraival_Place.append(A_P.text)
        
        stops = i.find_all("span", class_="VG3hNb")
        if stops:
            stop.append([s.text for s in stops])
        
        times = i.find("div",class_="gvkrdb AdWm1c tPgKwe ogfYpf")
        if times:
            time.append(times.text)
        
        weights = i.find("div", class_="O7CXue")
        if weights:
            lag.append(weights.text)
        
        cls = i.find("div", class_="cQnuXe k0gFV")
        if cls:
            classes.append(cls.text)
```

```python
df1 = pd.DataFrame({"Name":N,
                    "Cabin":["Economy"]*len(N),
                    "Departure Time":depart, 
                    "Arrival Time":arraival, 
                    "Departure Place":Depature_Place,
                   "Arraival Place":Arraival_Place,
                   "Total Travel Time":time,
                   "Stops":stop,
                    "Laguage Weight":lag,
                  "Prices":prices
                   })
```
